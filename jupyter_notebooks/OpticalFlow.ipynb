```python
import cv2 as cv
import numpy as np

cap = cv.VideoCapture('SEQ_0516.wmv')

if (cap.isOpened()== False): 
    print("Error opening video stream or file")  

imageSize = (320, 240)
dispSize = (640, 480)

ret, frame1 = cap.read()
frame1 = cv.resize(frame1, imageSize)
prvs = cv.cvtColor(frame1,cv.COLOR_BGR2GRAY)
prevDiff = cv.absdiff(prvs, prvs)
hsv = np.zeros_like(frame1)
hsv[...,1] = 255
hsv[...,0] = 255

while(1):
    ret, frame2 = cap.read()
    
    if ret == True:
        frame2 = cv.resize(frame2, imageSize)
        nextF = cv.cvtColor(frame2,cv.COLOR_BGR2GRAY)
        
        thisDiff = cv.absdiff(prvs, nextF)
                       
        flow = cv.calcOpticalFlowFarneback(prevDiff, thisDiff, None, 0.5, 3, 15, 3, 5, 1.2, 0)
        mag, ang = cv.cartToPolar(flow[...,0], flow[...,1])
        hsv[...,0] = ang*180/np.pi/2
        hsv[...,2] = cv.normalize(mag,None,0,255,cv.NORM_MINMAX)
        bgr = cv.cvtColor(hsv,cv.COLOR_HSV2BGR)
        bgr = cv.resize(bgr, (640, 480))
        
        vectorized = bgr.reshape((-1,3))
        vectorized = np.float32(vectorized)
        criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 5, 1.0)
        K = 2
        attempts=5
        ret,label,center=cv.kmeans(vectorized,K,None,criteria,attempts,cv.KMEANS_PP_CENTERS)
        center = np.uint8(center)
        res = center[label.flatten()]
        result_image = res.reshape((bgr.shape))
        
        if np.shape(result_image) != ():
            result_image = cv.resize(result_image, dispSize)
            cv.imshow('frame2',result_image)
        if np.shape(frame2) != ():
            frame2 = cv.resize(frame2, dispSize)
            cv.imshow('VideoStream',frame2)
        if np.shape(bgr) != ():
            bgr = cv.resize(bgr, dispSize)
            cv.imshow('bgr',bgr)
        if np.shape(thisDiff) != ():
            diff = cv.resize(thisDiff, dispSize)
            cv.imshow('diff',diff)
        if cv.waitKey(10) & 0xFF == ord('q'):
            break 
    else: 
        break
    prvs = nextF
    prevDiff = thisDiff

cap.release() 
cv.destroyAllWindows()
```
